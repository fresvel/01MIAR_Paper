### Resumen del Artículo

El estudio analizado [2] se centra en dos librerías populares de Python para el análisis y manipulación de datos: Pandas y Polars. A continuación se define y detalla algunas características mencionadas en el artículo, para contextualizar estas dos librerías:

**Pandas:** Es una librería de Python ampliamente utilizada para trabajar con conjuntos de datos. Ofrece funciones para analizar, limpiar, explorar y manipular datos. Simplifica la carga de datos de diversos formatos (CSV, Excel, y DBMS) y asiste en la limpieza de datos, la exploración con herramientas estadísticas y de visualización, y la transformación de datos[2].

**Polars:** Es una librería de manipulación de datos basada en Rust, que proporciona capacidades de procesamiento de datos similares a Pandas, pero con un enfoque en el rendimiento. Polars se integra con proyectos Python a través de sus Python bindings y con librerías comunes de Python como NumPy y PyArrow. Se diferencia por el uso de la evaluación perezosa (lazy evaluation), donde las operaciones construyen un plan de cómputo en lugar de ejecutarse inmediatamente, esto mejora la eficiencia en operaciones complejas. También aprovecha el poder de los procesadores multinúcleo para ejecutar operaciones de datos concurrentemente, acelerando las tareas de procesamiento.


El estudio reconoce que, aunque ya existían benchmarks que comparaban el rendimiento de Polars con otras soluciones (incluyendo Pandas) y reportaban una mejor velocidad para Polars, estos no eran sistemáticos ni revisados por pares. La contribución específica de esta investigación radica en estudiar el impacto energético de usar Pandas y Polars en diferentes escenarios, como el tamaño del dataset y las operaciones realizadas, para determinar si la mejora en rendimiento de Polars se traduce también en un menor consumo de energía.

```python
# Importar las librerías necesarias
import pandas as pd
import polars as pl

# --- Cargar datos desde un archivo CSV en un DataFrame ---

# Usando Pandas
print("Cargando datos con Pandas...")
pd_df = pd.read_csv('data.csv') # Carga un archivo CSV en un DataFrame de Pandas [18]
print("DataFrame de Pandas cargado. Primeras 5 filas:")
print(pd_df.head()) # Muestra las primeras 5 filas del DataFrame de Pandas

# Usando Polars
print("\nCargando datos con Polars...")
pl_df = pl.read_csv('data.csv') # Carga un archivo CSV en un DataFrame de Polars [18]
print("DataFrame de Polars cargado. Primeras 5 filas:")
print(pl_df.head()) # Muestra las primeras 5 filas del DataFrame de Polars


# --- Filtrar filas donde column_name es mayor que 50 ---
column_name='network_packet_size'
# Usando Pandas
print("\nFiltrando datos con Pandas...")
# Se asume que column_name existe en el DataFrame 'pd_df'
pd_filtered_df = pd_df[pd_df[column_name] > 50] # Filtra filas donde column_name es > 50 [18]
print("DataFrame de Pandas filtrado. Número de filas:", len(pd_filtered_df))

# Usando Polars
print("\nFiltrando datos con Polars...")
# Se asume que column_name existe en el DataFrame 'pl_df'
pl_filtered_df = pl_df.filter(pl.col(column_name) > 50) # Filtra filas donde column_name es > 50 (se usa pl.col para seleccionar la columna) [18]
print("DataFrame de Polars filtrado. Número de filas:", len(pl_filtered_df))


# --- Ordenar el DataFrame por column_name ---
column_name='session_duration'
# Usando Pandas
print("\nOrdenando datos con Pandas...")
# Se asume que column_name existe en el DataFrame 'pd_df'
pd_sorted_df = pd_df.sort_values(by=column_name) # Ordena el DataFrame por column_name [18]
print("DataFrame de Pandas ordenado. Primeras 5 filas de la columna column_name:")
print(pd_sorted_df[column_name].head())

# Usando Polars
print("\nOrdenando datos con Polars...")
# Se asume que column_name existe en el DataFrame 'pl_df'
pl_sorted_df = pl_df.sort(column_name) # Ordena el DataFrame por column_name [18]
print("DataFrame de Polars ordenado. Primeras 5 filas de la columna column_name:")
print(pl_sorted_df.select(column_name).head()) # Selecciona y muestra las primeras 5 filas de la columna column_name
```

En el siguiente código se realiza una gráfica de los tiempos que le toma a cada librería cargar el csv, filtrar los datos y ordenar los mismos.

```python
import pandas as pd
import polars as pl
import time
import matplotlib.pyplot as plt

# Configuración
csv_file = 'data.csv'
column_filter = 'network_packet_size'
column_sort = 'session_duration'

# Diccionario para almacenar los tiempos
timings = {
    'Cargar CSV': {'Pandas': 0, 'Polars': 0},
    'Filtrar > 50': {'Pandas': 0, 'Polars': 0},
    'Ordenar': {'Pandas': 0, 'Polars': 0}
}

# --- Pandas ---
print("\n--- Medición con Pandas ---")

start = time.time()
pd_df = pd.read_csv(csv_file)
timings['Cargar CSV']['Pandas'] = time.time() - start

start = time.time()
pd_filtered = pd_df[pd_df[column_filter] > 50]
timings['Filtrar > 50']['Pandas'] = time.time() - start

start = time.time()
pd_sorted = pd_df.sort_values(by=column_sort)
timings['Ordenar']['Pandas'] = time.time() - start

# --- Polars ---
print("\n--- Medición con Polars ---")

start = time.time()
pl_df = pl.read_csv(csv_file)
timings['Cargar CSV']['Polars'] = time.time() - start

start = time.time()
pl_filtered = pl_df.filter(pl.col(column_filter) > 50)
timings['Filtrar > 50']['Polars'] = time.time() - start

start = time.time()
pl_sorted = pl_df.sort(column_sort)
timings['Ordenar']['Polars'] = time.time() - start

# --- Gráfica comparativa ---
etapas = list(timings.keys())
pandas_times = [timings[etapa]['Pandas'] for etapa in etapas]
polars_times = [timings[etapa]['Polars'] for etapa in etapas]

x = range(len(etapas))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar(x, pandas_times, width=width, label='Pandas', align='center')
plt.bar([i + width for i in x], polars_times, width=width, label='Polars', align='center')
plt.xticks([i + width / 2 for i in x], etapas)
plt.ylabel('Tiempo (segundos)')
plt.title('Comparación de tiempo: Pandas vs Polars')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

### Resultados

#### Restricciones y Limitaciones

En el artículo se puede identificar las siguientes restricciones y limitaciones para Pandas y Polars:

**Limitaciones de Pandas**

**Mayor consumo de energía:** Pandas consume considerablemente más energía que Polars, especialmente al manipular grandes dataframes. En el bloque de DATs sintéticos, Pandas consumió aproximadamente 8 veces más energía que Polars para grandes dataframes. Para los DATs del Benchmark TPCH de gran tamaño, Pandas necesitó aproximadamente el 63% más de energía que Polars.

**Menor eficiencia energética en general:** Las medias de consumo de energía de Polars fueron consistentemente más bajas que las de Pandas en casi todos los bloques de estudio. Para dataframes pequeños en los DATs sintéticos, Polars también mostró un rendimiento significativamente mejor en términos de energía.

**Mayor tiempo de ejecución:** El tiempo de ejecución promedio es más rápido para Polars, especialmente para dataframes grandes. Las operaciones de datos en Pandas son "intensivas en recursos y pueden consumir mucha potencia computacional".

**Correlación positiva con el uso de memoria:** Pandas exhibió una correlación positiva estadísticamente entre el uso de energía y el uso de memoria en todos los bloques del experimento. Esto sugiere que un mayor uso de memoria en Pandas se traduce directamente en un mayor consumo de energía. Además, la literatura gris sugiere que Pandas puede experimentar errores de memoria insuficiente (out-of-memory) más fácilmente que Polars al manejar dataframes grandes.

**Procesamiento de un solo hilo (por defecto):** Pandas, por defecto, utiliza un solo núcleo de CPU
. Esto es una limitación importante, ya que no aprovecha los procesadores multinúcleo modernos para el procesamiento paralelo.

**Correlación negativa con el uso de CPU:** Se observó una correlación negativa estadísticamente significativa entre el uso de energía y el uso de CPU para Pandas en todos los bloques del experimento. Esto es una consecuencia de su utilización de un solo núcleo, lo que sugiere que incluso con mayor utilización de CPU, la eficiencia energética disminuye o no mejora como se esperaría en sistemas multinúcleo.

**Evaluación eager:** Pandas emplea la evaluación eager (ávida), lo que significa que las operaciones se ejecutan inmediatamente. Esto puede llevar a planes de cálculo menos optimizados para operaciones de datos complejas en comparación con la evaluación lazy de Polars.

**Rendimiento en operaciones de E/S:** En el segmento de DATs sintéticos, Pandas mostró un rendimiento notablemente inferior, atribuido principalmente a las operaciones de Entrada/Salida (E/S) con datos. Esto fue más evidente al comparar con el Benchmark TPCH, que probablemente tenía menos DATs de esta naturaleza y utilizaba archivos Parquet de menor tamaño y más rápida recuperación.

**Mayor variabilidad en el uso de energía:** Para dataframes grandes, la desviación estándar del uso de energía de Pandas fue 3 veces mayor para el dataframe TPCH grande y casi 5 veces mayor para el dataframe sintético grande, en comparación con Polars. Esto indica una mayor inestabilidad y menos predictibilidad en su consumo de energía.

**Limitaciones de Polars**

**Madurez y funcionalidad:** Polars es una biblioteca relativamente nueva y aún carece de algunas de las funcionalidades que Pandas ya ofrece. 

**Diferencia de energía para dataframes TPCH pequeños:** Para dataframes pequeños en los DATs del Benchmark TPCH, el estudio no pudo encontrar una diferencia estadísticamente significativa en el uso de energía promedio entre Pandas y Polars. Esto indica que Polars no siempre ofrece una ventaja energética significativa en todos los escenarios de conjuntos de datos pequeños.

**Posible correlación memoria-energía para dataframes TPCH grandes:** Aunque generalmente Polars no mostró una correlación significativa entre el uso de energía y memoria, se observó una correlación fuerte (0.71) y estadísticamente significativa para Polars específicamente dentro del bloque de experimentos TPCH para dataframes grandes. Esto sugiere que, en ciertos contextos de tareas y tamaños de datos, la gestión de memoria de Polars aún podría influir en su consumo de energía, aunque su coeficiente de correlación fue consistentemente menor que el de Pandas en la mayoría de las pruebas.

#### Mejoras de Polars

Las mejoras de Polars, en comparación con Pandas, según el artículo [2], se centran de manera principal en su eficiencia energética y rendimiento:

**Mayor eficiencia energética:** Polars es más eficiente energéticamente que Pandas, especialmente al manipular grandes dataframes.

** Menor tiempo de ejecución y mayor velocidad:** Polars es más rápido en el tiempo de ejecución promedio, especialmente para dataframes grandes. Valida la afirmación de los mantenedores de Polars de que tiene un tiempo de ejecución mucho más corto que su competencia, especialmente Pandas.

**Múltiples núcleos de CPU (procesamiento paralelo):** Polars aprovecha el poder de los procesadores multinúcleo para ejecutar operaciones de datos concurrentemente, acelerando las tareas de procesamiento. Esto contrasta con Pandas, que por defecto utiliza un solo núcleo de CPU. La falta de correlación entre el uso de energía y el uso de CPU en Polars se atribuye a su efectiva utilización de todos los núcleos de CPU.

**Gestión de memoria más eficiente:** Polars parece emplear una gestión de memoria más eficiente. La correlación entre el uso de energía y el uso de memoria es consistentemente menor para Polars en la mayoría de las pruebas en comparación con Pandas. Aunque se observó una correlación positiva y significativa para Polars en el bloque de TPCH con dataframes grandes, en general, las fuentes sugieren que Polars es menos propenso a errores de falta de memoria (out-of-memory) que Pandas al manejar dataframes grandes.

**Implementación basada en Rust:** Polars está implementado en Rust, un lenguaje conocido por su rendimiento y seguridad de memoria. Esto contribuye a su eficiencia general.

La utilización de los núcleos de CPU es un factor crucial en la selección de librerías para el análisis de datos, ya que Polars aprovecha por defecto todos los núcleos disponibles, mientras que Pandas utiliza uno solo.

Basándose en estos hallazgos, los autores de [2] recomiendan a los científicos de datos utilizar Polars para tareas de análisis de datos que requieran velocidad y eficiencia energética, especialmente cuando se trabaja con dataframes grandes. A pesar de que Polars es una librería más reciente y aún carece de algunas funcionalidades que Pandas ya ofrece, sus ventajas en rendimiento y eficiencia energética son evidentes. El estudio sugiere que los desarrolladores de Pandas podrían considerar implementar el uso de múltiples núcleos de CPU.
